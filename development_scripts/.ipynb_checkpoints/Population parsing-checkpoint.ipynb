```python
import xml
```

```python
model_labels = ['user_notes', 'production_type', 'latitude', 'longitude', 'initial_state', 'initial_size']
xml_fields   = ['id',             'production-type', 'latitude', 'longitude', 'status',        'size']
text_fields = list(zip(model_labels, xml_fields))
text_fields
```

```python
def getText(elem):
    return ",".join(elem.itertext())
```

```python
from xml.etree.ElementTree import ElementTree
class PopulationParser:
    def __init__(self, filename):
        tree = ElementTree(file='' + filename)
        self.top_level = tree.getroot()
        self.population = []
        
    def parse_to_dictionary(self):
        self.parse_text_fields(text_fields)
        return self.population
        
    def parse_text_fields(self, text_fields):
        for herd in self.top_level.iter('herd'):
            self.population.append( dict() ) #empty
            for t in text_fields:
                if isinstance(t, tuple):
                    self.populate_text_field(herd, *t)
                else:
                    self.populate_text_field(herd, t)
                
    def populate_text_field(self, herd, field_name, anathema_name = ''):
        if not anathema_name:
            anathema_name = field_name
        text = ''
        try:
            element = next(herd.iter(anathema_name))
            text = getText(element)
        except: 
            raise IOError("Couldn't find '%s' label in xml" % anathema_name)
        self.population[-1][field_name] = text
```

```python
parser = PopulationParser('Population_Grid.xml')
```

```python
choices=(('L', 'Latent'),
                                              ('S', 'Susceptible'),
                                              ('B', 'Subclinical'),
                                              ('C', 'Clinical'),
                                              ('N', 'Naturally Immune'),
                                              ('V', 'Vaccine Immune'),
                                              ('D', 'Destroyed'))
```

```python
def squish_name(name):
    return name.lower().strip().replace(' ', '')
```

```python
squish_name('  The FeanCiest Evar  ')
```

```python
def value_to_db_repr(value, map_tuple):
    value = squish_name(value)
    for key, full_str in map_tuple:
        if value == squish_name(full_str):
            return key
    return 'L'
```

```python
tests = ['Vaccine immune', 'destroyed', '  Latent', 'SubClinical', 'kittens']
for x in tests:
    print( value_to_db_repr(x, choices) )
```
