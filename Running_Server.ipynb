```python
field_name = 'tstcUAll'
for pt in list(ProductionType.objects.all()) + [None]:
    for iteration in list_of_iterations():
        print("For", pt)
        tstCUAll = {day: value for day, value in list(DailyByProductionType.objects.filter(iteration=iteration, production_type=pt).order_by('day').values_list('day', field_name))}
        for day in range(1, len(tstCUAll)):
            current = DailyByProductionType.objects.filter(day=day, iteration=iteration, production_type=pt).first()
            previous = 0
            try:  previous = tstCUAll[day-1]
            except: print('Fail')
            current.tstnUAll = current.tstcUAll - previous
            current.save()
            print(current.tstnUAll)
        print('\n')
```
